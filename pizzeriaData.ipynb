```python
def live_data(url):
    data = requests.get(url)
    return data
```

```python
df_pizza_outlets.reset_index(inplace=True)
df_pizza_outlets
```

```python
df_pizza_outlets = df_pizza_outlets[["address","categories","city","menus.name","name","menus.amountMax","menus.amountMin"]]
df_pizza_outlets
```

##### Only keeping the necessary columns that are required for the challenge:
##### address, categories, city, menus.description, menus.name, name, menus.amountMax, menus.amountMin

```python
df_pizza_outlets_city = df_pizza_outlets[["city","name","menus.amountMax","menus.amountMin","menus.name"]]
```

```python
df_pizza_outlets_city = df_pizza_outlets.groupby(by='city')[['name']].nunique()
df_pizza_outlets_city.reset_index(inplace=True)
df_pizza_outlets_city.sort_values(by = 'name',ascending = False)
```

##### name now shows the amount of pizzerias in a city within Pennsylvania

```python
df_city_pop.rename(columns = {"City":"city"},inplace=True)
df_city_pop
```

```python
merged_df = pd.merge(df_pizza_outlets_city,df_city_pop,how="left",on='city')
```

##### Renaming city columns to have the same name, to merge the datasets so that population data is available on the pizzeria dataset

```python
merged_df.sort_values(by = 'name',ascending = False)
```

##### By sorting name It will show the cities with the greatest amount of pizzerias and also show  their population's size

```python
merged_df["Population"].fillna(1,inplace=True)
```

```python
merged_df.info()
```

```python
merged_df['Population'] = merged_df.apply(lambda x: int(x["Population"]),axis=1)
```

```python
merged_df['Pizzeria_density'] = merged_df.apply(lambda x: x["Population"]/x["name"],axis=1)
merged_df
```

##### Works out the amount of people per pizzeria in each country. This can be used to determine in which cities are pizzerias the most popular or where there are sparsely located, which could potentially mean less competition.

```python
merged_df.sort_values(by = 'Pizzeria_density',ascending = False)
```

```python
merged_df.nlargest(15,["Pizzeria_density"])
```

```python
shortlist_df = merged_df[merged_df["city"].isin(['Philadelphia','Pittsburgh','Scranton'])]
shortlist_df
```

##### Worked out the sum of every mentioned meal item to decipher which are most popular

```python
menu_items = df_pizza_outlets[["menus.name","menus.amountMax"]].groupby(by="menus.name").agg(["count","max"])
menu_items.reset_index(inplace=True)
menu_items
```

```python
menu_items.columns = menu_items.columns.droplevel()
```

```python
menu_items.sort_values(by="count",ascending=False,inplace=True)
menu_items.head(60)
```
